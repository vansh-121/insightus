```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
df = pd.read_csv('/content/heart_disease_uci.csv')
```

```python
df.count()
```

```python
df.isnull().any()
```

```python
for column in df.columns:
  if df[column].dtype != 'object': # Check if the column is not boolean
      # Plot a histogram
    plt.hist(df[column], bins=5, edgecolor='black')

    # Customize the plot
    plt.title('Histogram of Values')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Display the plot
    plt.show()
  else: # Handle boolean columns differently
    # Count the occurrences of True and False
    value_counts = df[column].value_counts()

    # Plot a bar chart
    plt.bar(value_counts.index, value_counts.values)

    # Customize the plot
    plt.title('Distribution of Boolean Values')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks([0, 1], ['False', 'True'])  # Set x-axis labels

    # Display the plot
    plt.show()
```
